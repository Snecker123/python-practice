```python
# End to End Example with Pandas & Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create CSV
data = {
    'Name': ['Gabriel', 'Vali', 'Madalina', 'Tudor', 'Andrei'],
    'Department': ['IT', 'HR', 'IT', 'HR', 'Marketing'],
    'Age': [25, 32, 28, 31, None],
    'Salary': [4000, 3500, 4200, 3300, 3900]
}

df = pd.DataFrame(data)
df.to_csv('employees.csv', index=False)

print("CSV saved as employees.csv")
```

```python
# Load CSV
df = pd.read_csv('old/employees.csv')
df
```

```python
# Explore
print("First rows:")
print(df.head())
print("\\nInfo:")
print(df.info())
print("\\nDescribe:")
print(df.describe())
```

```python
# Clean data
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
# Group
mean_salaries = df.groupby('Department')['Salary'].mean()
mean_salaries
```

```python
# Plot
mean_salaries.plot(kind='bar', color='coral', title='Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary (RON)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
